```python
#Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
```

```python
pies = ["Apple", "Pumpkin", "Chocolate Creme", "Cherry", "Apple Crumb", "Pecan", "Lemon Meringue", "Blueberry", "Key Lime", "Peach"]
```

```python
pie_votes = [47,37,32,27,25,24,24,21,18,16]
```

```python
plt.pie(pie_votes)
plt.show()
```

```python
plt.pie(pie_votes,labels=pies)
plt.show()
```

```python
plt.pie(pie_votes,labels=pies,autopct = "%1.1f%%")
plt.show()
```

```python
explode_values = (0, 0, 0, 0.2, 0, 0, 0, 0, 0,0)
plt.pie(pie_votes, explode=explode_values, labels=pies, autopct='%.1f%%')
```

```python
colors = ["yellow","green","lightblue","orange","red","purple","pink","yellowgreen","lightskyblue","lightcoral"]
explode_values = (0, 0, 0, 0.2, 0, 0, 0, 0, 0,0)
plt.pie(pie_votes, explode=explode_values,colors=colors, labels=pies, autopct='%.1f%%')
```
